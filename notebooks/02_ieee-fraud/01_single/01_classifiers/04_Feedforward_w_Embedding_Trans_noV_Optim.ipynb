# Changing the Optimizer
---
Throughout the notebooks we have been using the `AdamW` Optimizer, because it is often a good place to begin and according to most literature it tends to converge faster. 

Now, the same sort of literature suggests that sometimes standard `SGD` w/Momentum converges to a more optimal solution. But `SGD` is much less forgiving to incorrect hyperparameters like the learning rate, it tends to take more time to tune.

In this notebook, we're going to try and see what happens if we change Optimizers.

````python
with en.EnginePandas(num_threads=1) as e:
    ti = e.np_from_csv(
        (td_categorical, td_counts, td_timing, td_match, td_label), 
        tr_train_file, inference=False
    )
    
train, val, test = ti.split_sequential(50000,0)

train_sampler = mp.TensorInstanceNumpyLabelSampler().over_sampler(train)

train_ds = mp.TensorInstanceNumpyDataSet(train)
val_ds   = mp.TensorInstanceNumpyDataSet(val)

train_dl = train_ds.data_loader(torch.device('cpu'), 128, num_workers=2, sampler=train_sampler)
val_dl   = val_ds.data_loader(torch.device('cpu'), 128, num_workers=2)

mc = mp.ModelConfiguration.from_tensor_definitions(ti.target_tensor_def)
````

## Regular Adam

### Create Model

````python
class IEEEFeedForwardModel(mp.BinaryClassifier):
    def __init__(self, model_configuration: mp.ModelConfiguration):
        # Always call init, it is very important to correctly initialize the nn.Module
        super(IEEEFeedForwardModel, self).__init__(model_configuration)
        
        # Create the heads, they are the first layer of the NN.
        self.heads = self.create_heads(max_dims=30)
        heads_size = sum([h.output_size for h in self.heads])
        
        # Set of layers
        self.layers = mp.LinLayer(input_size=heads_size, layer_sizes=(128,64,64,32), dropout=0.0, bn_interval=2)
        
        # And a tail layer, for a binary classifier this is a Sigmoid.
        self.tail = self.create_tail(self.layers.output_size)

    # Create a forward-pass function through the neural net. Make sure to return a Tuple of Tensors!
    def forward(self, x: Tuple[torch.Tensor, ...]) -> Tuple[torch.Tensor, ...]:
        o = torch.cat([h(x[i]) for i, h in enumerate(self.heads)], dim=1)
        o = self.layers(o)
        o = self.tail(o)
        return (o,)

#Create a 'IEEEFeedForwardModel' instance
model = IEEEFeedForwardModel(mc)
model
````

### Train Adam
Compared to the AdamW optimizer, the optimal learning curve seems to drop slightly sooner, we will use an ever so slighly lower learning rate of **2e-3**

````python
min_lr = 1e-5
max_lr = 1e-1
m_lr = IEEEFeedForwardModel(mp.ModelConfiguration.from_tensor_definitions(ti.target_tensor_def))
o_lr = mp.AdamOptimizer(m_lr, lr=min_lr)
t_lr = mp.Trainer(m_lr, data_device, train_dl, val_dl)
lrs  = t_lr.find_lr(o_lr, max_lr, 300)
mplt.TrainPlot.plot_lr(lrs);
del m_lr, o_lr, t_lr, lrs
_ = gc.collect()
````

````python
EPOCHS = 5

# First create a trainer.
trainer = mp.Trainer(model, model_device, train_dl, val_dl)

# Then an optimizer
optimizer = mp.AdamOptimizer(model, lr=2e-3)

# Set-up a OneCycle Scheduler
scheduler = torch.optim.lr_scheduler.OneCycleLR(
            optimizer=optimizer.optimizer, 
            max_lr=optimizer.lr,
            steps_per_epoch=trainer.number_of_train_steps,
            epochs=EPOCHS
)

# Run the trainer
history = trainer.train(EPOCHS, optimizer, scheduler)
````

````python
mplt.TrainPlot.plot_history(history, fig_size=(8,8));
````

### Test Adam
This is an interesting result, even though after training we can see that the validation accuracy seems to be lower than in our base model, we do get roughly the same AUC score.

There is an important difference however. If we compare to the base model, this model seems to be better at finding the fraud, it has a lower FN count. On the other hand the FP count is significantly higher. Potentially this is a sign that our base AdamW model converged faster, but might start overfitting to the fraud records. (Remember that we over sample quite dramatically)

````python
t = mp.Tester(model, torch.device(model_device), val_dl)
results = t.test_results_binary()
mplt.TestPlot.print_binary_classification_report(results)
mplt.TestPlot.plot_binary_confusion_matrix(results, fig_size=(5,5))
mplt.TestPlot.plot_roc_curve(results, fig_size=(6,6))
mplt.TestPlot.plot_precision_recall_curve(results, fig_size=(6,6))
````

## SGD w/Momentum

### Create Model

````python
class IEEEFeedForwardModel(mp.BinaryClassifier):
    def __init__(self, model_configuration: mp.ModelConfiguration):
        # Always call init, it is very important to correctly initialize the nn.Module
        super(IEEEFeedForwardModel, self).__init__(model_configuration)
        
        # Create the heads, they are the first layer of the NN.
        self.heads = self.create_heads(max_dims=30)
        heads_size = sum([h.output_size for h in self.heads])
        
        # Set of layers
        self.layers = mp.LinLayer(input_size=heads_size, layer_sizes=(128,64,64,32), dropout=0.0, bn_interval=2)
        
        # And a tail layer, for a binary classifier this is a Sigmoid.
        self.tail = self.create_tail(self.layers.output_size)

    # Create a forward-pass function through the neural net. Make sure to return a Tuple of Tensors!
    def forward(self, x: Tuple[torch.Tensor, ...]) -> Tuple[torch.Tensor, ...]:
        o = torch.cat([h(x[i]) for i, h in enumerate(self.heads)], dim=1)
        o = self.layers(o)
        o = self.tail(o)
        return (o,)

#Create a 'IEEEFeedForwardModel' instance
model = IEEEFeedForwardModel(mc)
model
````

### Train SGD

````python
min_lr = 1e-5
max_lr = 1e-1
m_lr = IEEEFeedForwardModel(mp.ModelConfiguration.from_tensor_definitions(ti.target_tensor_def))
o_lr = mp.SGDOptimizer(m_lr, lr=min_lr, momentum=0.90)
t_lr = mp.Trainer(m_lr, data_device, train_dl, val_dl)
lrs  = t_lr.find_lr(o_lr, max_lr, 300)
mplt.TrainPlot.plot_lr(lrs);
del m_lr, o_lr, t_lr, lrs
_ = gc.collect()
````

````python
EPOCHS = 5

# First create a trainer.
trainer = mp.Trainer(model, model_device, train_dl, val_dl)

# Then an optimizer
optimizer = mp.SGDOptimizer(model, lr=5e-3, momentum=0.90)

# Set-up a OneCycle Scheduler
scheduler = torch.optim.lr_scheduler.OneCycleLR(
            optimizer=optimizer.optimizer, 
            max_lr=optimizer.lr,
            steps_per_epoch=trainer.number_of_train_steps,
            epochs=EPOCHS
)

# Run the trainer
history = trainer.train(EPOCHS, optimizer, scheduler)
````

````python
mplt.TrainPlot.plot_history(history, fig_size=(8,8));
````

### Test SGD


````python
t = mp.Tester(model, torch.device(model_device), val_dl)
results = t.test_results_binary()
mplt.TestPlot.print_binary_classification_report(results)
mplt.TestPlot.plot_binary_confusion_matrix(results, fig_size=(5,5))
mplt.TestPlot.plot_roc_curve(results, fig_size=(6,6))
mplt.TestPlot.plot_precision_recall_curve(results, fig_size=(6,6))
````

## Conclusion
It is possible we ran into a fluke, accidently got a good a set of good parameters, but it does feel like for this model and data, the SGD w/momentum and our dynamic learning rate slightly outperforms the other optimizers. Moreover from the literature, if it is true that SGD requires more training and hyper parameter tweaking, it feels like there is some more potential to get the SGD to work better. 

From some initial experiments it seems like the SGD get better with one or two extra epochs. This is not the case for Adam and AdamW.
